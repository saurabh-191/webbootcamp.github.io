First we have to import important module which were use in this project.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
flights = pd.read_csv('/content/flights_data.csv')
print(flights.shape)
flights.head()
```

Now we have to print the Bar chat for that dataset.

```python
#code for the Bar Chat
sb.countplot(data = flights, x = 'Source')

plt.ylabel('Number of Flights', fontsize = 12)
plt.xlabel('Source', fontsize = 12)
```

```python
base_color = sb.color_palette()[0]
sb.countplot(data= flights , x = 'Source', color = base_color)
plt.xticks(rotation=40);
```

```python
base_color = sb.color_palette()[1]
gen_order = flights['Source'].value_counts().index
sb.countplot(data =flights, x= 'Source', color=base_color,
             order = gen_order)
```

```python
 base_color = sb.color_palette()[2]
 sb.countplot(data=flights, x= 'Airline', color = base_color)
```

```python
base_color = sb.color_palette()[2]
sb.countplot(data=flights, x= 'Airline', color = base_color)
plt.xticks(rotation = 90)
```

```python
flights.isna().sum()
```

```python
fa_count = flights.isna().sum()
base_color = sb.color_palette()[0]
sb.barplot(fa_count.index.values, fa_count, color = base_color)
plt.xticks(rotation= 90)
plt.ylabel('Number of missing values', fontsize=12);
```

Now we have to made pie char for the data set

```python
sorted_count = flights['Destination'].value_counts()
plt.pie(sorted_count, labels = sorted_count.index, startangle = 90,
         counterclock = False);
plt.axis('square')
plt.title('Flight Destination')
```

```python
sorted_count = flights['Destination'].value_counts()
plt.pie(sorted_count, labels = sorted_count.index, startangle = 90,
         counterclock = False, wedgeprops ={'width' : 0.4});
plt.axis('square');
```

```python
plt.hist(data = flights, x = 'Duration(minutes)')
```

```python
plt.hist(data = flights, x = 'Price', bins = 30)
```

```python
bins = np.arange(0, flights['Price'].max()+1, 1200)
plt.hist(data = flights, x = 'Price', bins = bins)
plt.show()
```

```python
sb.distplot(flights['Price'])
```

```python
sb.distplot(flights['Price'], kde = False)
```

```python
bin_edge = np.arange(0, flights['Price'].max()+1, 1000)
sb.distplot(flights['Price'], bins = bin_edge, kde = False,
            hist_kws ={'alpha':1})
```
